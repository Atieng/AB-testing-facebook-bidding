#### **Column Descriptions**

1. **Timestamp**: The date and time the survey response was submitted.
2. **Department**: The department (e.g., Fiber) where the respondent works.
3. **Role/Position**: The position of the respondent (e.g., Technician, Supervisor).
4. **Subsidiary**: The subsidiary (e.g., Kenya, Uganda) where the respondent is based.

5. **How often do you receive safety training?**: Frequency of safety training for the respondent (e.g., Quarterly, Monthly).

6. **The safety training I receive is relevant and useful.**: Level of agreement with whether the safety training is useful.
7. **I am aware of the safety policies and procedures applicable to my job**.: Level of agreement with awareness of job-specific safety policies.

8. **I feel comfortable reporting safety hazards or concerns.**: Level of agreement with comfort in reporting hazards.
9. **Management takes safety concerns seriously and addresses them promptly.**: Level of agreement with management's response to safety concerns.
10. **I regularly use personal protective equipment (PPE) as required.**: Frequency of using PPE at work.
11. **I am encouraged to participate in safety meetings and discussions.**: Level of agreement with encouragement to join safety discussions.
12. **There is clear communication about safety updates and changes in procedures.**: Level of agreement with how well safety updates are communicated.
13. **I believe that safety is a top priority in our organization.**: Level of agreement with safety being a priority in the organization.
14. **I feel safe working in my current environment.**: Level of agreement with feeling safe at work.
15. **What suggestions do you have for improving safety in our organization?**: Open-ended feedback or suggestions for improving safety.
16. **Please share any safety incidents or near misses you think should be addressed**: Any reported safety incidents or near-misses.
17. **Please give the name of any person you may recognize for his/her commitment to safety**: Recognition of individuals committed to safety.

18. **Please give the reason for the above**: The reason why the person is being recognized for their safety commitment.
